```python
import pandas as pd
df = pd.read_csv("/content/winequalityN.csv")
df
```

```python
red_wine_df = df[df['type'] == 'red']
white_wine_df = df[df['type'] == 'white']
wine_combined_df = pd.concat([red_wine_df, white_wine_df], ignore_index=True)
wine_combined_df
```

```python
# Group the DataFrame by 'wine_type' and calculate the average alcohol content
average_alcohol_by_type = wine_combined_df.groupby('type')['alcohol'].mean()

# Print the average alcohol content for red and white wines
print("Average Alcohol Content for Red and White Wines:")
print(average_alcohol_by_type)
```

```python

import matplotlib.pyplot as plt

red_wine_quality = wine_combined_df[wine_combined_df['type'] == 'red']['quality']
white_wine_quality = wine_combined_df[wine_combined_df['type'] == 'white']['quality']

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.hist(red_wine_quality, bins=range(3, 9), edgecolor='k', alpha=0.7)
plt.title("Red Wine Quality")
plt.xlabel("Quality Score")
plt.ylabel("Frequency")

plt.subplot(1, 2, 2)
plt.hist(white_wine_quality, bins=range(3, 9), edgecolor='k', alpha=0.7)
plt.title("White Wine Quality")
plt.xlabel("Quality Score")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
# Calculate the correlation matrix
correlation_matrix = wine_combined_df.corr()

# Print the correlation matrix
print("Correlation Matrix between Wine Attributes:")
print(correlation_matrix)
```

```python

red_wine_df = wine_combined_df[wine_combined_df['type'] == 'red']
white_wine_df = wine_combined_df[wine_combined_df['type'] == 'white']

# Create scatter plots for red and white wines
plt.figure(figsize=(10, 6))

# Scatter plot for red wines
plt.scatter(red_wine_df['alcohol'], red_wine_df['quality'], color='red', alpha=0.5, label='Red Wine')

# Scatter plot for white wines
plt.scatter(white_wine_df['alcohol'], white_wine_df['quality'], color='white', alpha=0.5, label='White Wine')

plt.xlabel('Alcohol Content')
plt.ylabel('Quality')
plt.title('Relationship Between Alcohol Content and Quality for Red and White Wines')
plt.legend()
plt.grid(True)

# Show the scatter plot
plt.show()
```

```python


high_quality_wines = wine_combined_df[wine_combined_df['quality'] >= 7]


total_wines = len(wine_combined_df)


percentage_high_quality = (len(high_quality_wines) / total_wines) * 100

print(f"Percentage of High-Quality Wines (Quality Score >= 7): {percentage_high_quality:.2f}%")
```
